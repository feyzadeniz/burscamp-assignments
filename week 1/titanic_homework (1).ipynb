# ⚙️ ORTA SEVİYE SORULAR
---

### Soru 1: Cinsiyete göre yaş ortalamasını bulun.

```python
df.groupby("Sex")["Age"].mean()
```

### Soru 2: Sınıfa göre hayatta kalan kişi sayısını bulun.

```python
df.groupby("Pclass")["Survived"].count() # sonuna herhangi bir fonk. eklemediğim için <pandas.core.groupby.generic.SeriesGroupBy object at 0x7b062755f4d0> böyle bir sonuç almıştım, count veya sum ekleyerek devam etmem gerekiyormuş
```

### Soru 3: Hem cinsiyet hem sınıfa göre hayatta kalma oranlarını hesaplayın.

```python
df.groupby(["Sex", "Pclass"])["Survived"].agg(["mean", "median", "std"])
```

### Soru 4: Embarked değişkenine göre ortalama yaş ve ortalama ücret nedir?

```python
# df.groupby("Embarked")["Age", "Fare"].mean() -> başlangıçta böyle yazmıştım ve age, fare kısmına bir köşeli parantez daha eklemediğim için Pandas bunu tuple olarak algıladı. Doğrusu:
df.groupby("Embarked")[["Age", "Fare"]].mean().round(2) #daha düzenli görünmesi için round kullandım sayıları 2 ondalık basamağa yuvarladım
```

### Soru 5: Pclass gruplarına göre yaşın standart sapmasını bulun.

```python
df.groupby("Pclass")["Age"].agg(["mean","std"])
```

```python
oran = df.groupby("Pclass")["Age"].std() / df.groupby("Pclass")["Age"].mean()
print(oran)   # std yorumlamak için böyle bir hesaplama yazdım, sonuçlara baktığımızda standart sapmanın %30dan yüksek yani geniş bir aralıkta dağılmış olduğunu görüyorum. en yüksek sapmaya 3. sınıfta rastlıyoruz,
# bu sınıf içinde çok genç ve çok yaşlı bireyler bir arada olabilir.
```

### Soru 6: Her `Embarked` limanından kaç yolcu olduğunu ve bunların yüzdelik dağılımını yazdırın.

```python
df["Embarked"].value_counts()
```

```python
df["Embarked"].value_counts(normalize= True) * 100
```

### Soru 7: Yaşı 60'tan büyük olanların hayatta kalma oranını bulun.

```python
df[df["Age"] > 60]["Survived"].agg(["count","mean","std","median"]) # başlangıçta df.groupby(["Age"] > 60)["Survived"].agg(["mean","std","median"]) yazdım, hata aldım filtrelemek için tekrar başa df eklememiz gerekiyormuş.
```

### Soru 8: `Sex`, `Pclass`, `Survived` için `groupby()` + `agg()` ile ortalama yaş ve ücret hesaplayın.

```python
df.groupby(["Sex","Pclass","Survived"])[["Age", "Fare"]].mean()
```

### Soru 9: Age değişkenini artan düzende sıralayın ve ilk 10 kişiyi yazdırın.

```python
df.sort_values(by="Age", ascending=False).head(10)
```

### Soru 1: En pahalı bileti alan kişinin yaşı, cinsiyeti ve sınıfı nedir?

```python
df.loc[df["Fare"].idxmax(), ["Age", "Sex", "Pclass"]] # ben direkt  max ile  bulmayı denemiştim ama indeksini bulmamız gerekiyormuş, bu yüzden idxmax kullandık (en yüksek index)
```

### Soru 2: Hayatta kalanlar ile ölenler arasında yaş ortalamaları farkı nedir?

```python
ort = df.groupby("Survived")["Age"].mean()
```

```python
fark = abs(ort[1] - ort[0])
print(fark)  # abs negatif olmadan alır sonucu, 1 hayatta kalanlar 0 ölenler
```

### Soru 3: Hangi `Pclass` grubunda kadınların hayatta kalma oranı en yüksektir?

```python
kadınlar = df[df["Sex"] == "female"] # önce filtreledim, kadınlar için ayrı bir df oluşturdum
```

```python
oranlar = kadınlar.groupby("Pclass")["Survived"].mean() #kadınların pclass'a göre hayatta kalma oranları
print(oranlar)
```

```python
en_yuksek_sinif = oranlar.idxmax()
print(en_yuksek_sinif)
```

### Soru 4: Ortalama yaşın altında olan ve hayatta kalan erkeklerin sayısı kaçtır?

```python
ort_yas = df["Age"].mean()
```

```python
hayatta_erkek = df[(df["Age"]< ort_yas)&(df["Sex"] == "male")&(df["Survived"] == 1)]
```

```python
hayatta_erkek.shape[0]
```

### Soru 5: `Cabin` değeri eksik olmayanların yaş ortalaması ile eksik olanların yaş ortalaması arasında fark var mı?

```python
yas_cabin_var = df[df["Cabin"].notna()]["Age"].mean() # notnull ile denedim olmamıştı
```

```python
yas_cabin_yok = df[df["Cabin"].isna()]["Age"].mean()
```

```python
fark = abs(yas_cabin_var - yas_cabin_yok)
print(fark)
```
